```python
print(np.where(om['TOTAL']>350))
```

```python
Q1=om['TOTAL'].quantile(0.25)
Q3=om['TOTAL'].quantile(0.75)
IQR=Q3-Q1
Lower_limit=Q1 - 1.5*IQR
Upper_limit=Q1 + 1.5*IQR

print(f'Q1={Q1},Q3={Q3},IQR={IQR},LOWER LIMIT={Lower_limit}, UPPER LIMIT={Upper_limit}')
```

```python
for x in ['TOTAL']:
    q75,q25 = np.percentile(om.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    om.loc[om[x] < min,x] = np.nan
    om.loc[om[x] > max,x] = np.nan
```

```python
om.isnull().sum()
```

```python
om["TOTAL"]=om["TOTAL"].fillna(om["TOTAL"].mean())
```

```python
om.boxplot()
```

```python
for x in ['PERCENTAGE']:
    q75,q25 = np.percentile(om.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    om.loc[om[x] < min,x] = np.nan
    om.loc[om[x] > max,x] = np.nan
```

```python
om.isnull().sum()
```

```python
om["PERCENTAGE"]=om["PERCENTAGE"].fillna(om["PERCENTAGE"].mean())
```
